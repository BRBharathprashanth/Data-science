```python
## Password strenght checker

password = "Some@123123"

is_num_present = any(char.isdigit() for char in password)
is_capital_alpha_present = any(char.isupper() for char in password)
has_special_char = any(not char.isalnum() for char in password)

if len(password) >= 8 and is_num_present and is_capital_alpha_present and has_special_char:
    print("Password is strong enough!")
else:
    print("Password is not strong enough.")

print("Password check complete.")
```
