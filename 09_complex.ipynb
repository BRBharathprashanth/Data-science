```python
# Question 10: Find the phase angle of complex number 1 + 1j
import cmath
import math

z_q10 = 1 + 1j
phase_radians = cmath.phase(z_q10)
phase_degrees = math.degrees(phase_radians)

print(f"Complex number: {z_q10}")
print(f"Phase angle (radians): {phase_radians}")
print(f"Phase angle (degrees): {phase_degrees}")
```
