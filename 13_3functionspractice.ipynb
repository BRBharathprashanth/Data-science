```python
###1. Write a function that takes a number as input and returns whether it is even or odd.

def is_even_or_odd(number):
  if number % 2 ==0:
     return "even"
  else:
     return "odd"
```

```python
print(is_even_or_odd(28))
```

```python
##2. Write a function that takes a number and returns its factorial.

def factorial(number):
   if number ==0:
    return 1
   else:
    return number * factorial(number-1)
```

```python
print(factorial(5))
```

```python
##3. Write a function that takes a string and returns the reversed string.

def reverse_string(string):
    return string[::-1]
```

```python
print(reverse_string("prashanth"))
```

```python
##4. Write a function that checks whether a number is prime.

def is_prime_or_not(n):
  if n <=1:
      return False
  for i in range(2, int(n**0.5) + 1):
      if n % i == 0:
          return False
  return True
```

```python
print(is_prime_or_not(17))
```

```python
##5. Write a function that takes a list of numbers and returns the largest number.

def largest_number(n):
  if not n:
      return None # Handle empty list case
  largest = n[0]
  for num in n:
    if num > largest:
        largest = num # Corrected: should assign num, not n
  return largest
```

```python
print(largest_number([1, 12, 34, 89, 101]))
```

```python
##6. Write a function that takes a list and returns the sum of all elements without using the built-in sum() function.

def sum_list(numbers):
  total = 0
  for num in numbers:
      total += num
  return total
```

```python
print(sum_list([82]))
```

```python
##7. Write a function that counts the number of vowels in a string.

def count_vowels(s):
  vowels = "aeiouAEIOU"
  count = 0
  for char in s:
      if char in vowels:
         count += 1
  return count
```

```python
print(count_vowels("prashanth"))
```

```python
##8. Write a function that returns the Fibonacci sequence up to n terms.

def fibonacci(n):
    sequence = []
    a, b = 0, 1
    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b
    return sequence
```

```python
print(fibonacci(10))
```

```python
##9. Check whether a string is a palindrome

def is_palindrome(s):
    return s == s[::-1]
```

```python
print(is_palindrome("madam"))
```

```python
##10. Return the Greatest Common Divisor (GCD)

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a
```

```python
print(gcd(10,30))
```

```python
##11. Remove duplicates from a list

def remove_duplicates(lst):
    return list(set(lst))
```

```python
print(remove_duplicates([1, 2, 3, 3, 4, 4, 5]))
```

```python
##12. Return a list containing only even numbers

def get_even_numbers(numbers):
    return [num for num in numbers if num % 2 == 0]
```

```python
print(get_even_numbers([1,2,3,5,6,7,8,9]))
```

```python
##13. Swap two numbers without using a third variable

def swap_numbers(a, b):
    a, b = b, a
    return a, b
```

```python
print(swap_numbers(9, 10))
```

```python
##14. Convert Celsius to Fahrenheit

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
```

```python
print(celsius_to_fahrenheit(13))
```
