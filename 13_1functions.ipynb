```python
## calculate avg of three diff groups

group1_sum = 12 + 34 + 67
group1_avg = group1_sum / 3
print(f"Group 1 average: {group1_avg}")


group2_sum = 45 + 67 + 23
group2_avg = group2_sum / 3
print(f"Group 2 average: {group2_avg}")

group3_sum = 43 + 78 + 89
group3_avg = group3_sum / 3
print(f"Group 3 average: {group3_avg}")
```

```python
## With functions- Clear and REusable code

def calculate_average(scores):
    total = sum(scores)
    avg = total / len(scores)
    return avg

# use the fucntion

print(f'Groupd 1 average : {calculate_average([80,23,56])}')
```

```python
calculate_average([23,67,89,12])
```

```python
calculate_average([12,4,67,78,89,89,445,56,567])
```

```python
##function with multiple returns

def calculate(a, b):
  return a + b, a * b
sum_result, product_result = calculate(4, 2)
print(sum_result, product_result)
```

```python
def calculate(a, b):
  return a * b, a // b
product_result, float_result = calculate(10, 15)
print(product_result, float_result)
```

```python
##function with arbitary arguments(*args)

def add_all(*numbers):
    return sum(numbers)

print(add_all(1, 2, 3, 4))
```

```python
###Function with Keyword Arguments (**kwargs)

def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=25, city="New York")
```

```python
##lambda function(anonymous function)

square = lambda x: x ** 2
print(square(5))
```
