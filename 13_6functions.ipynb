```python
### lambda with single arguement

# square a number

square = lambda x : x**2

square(2)
```

```python
square(5)
```

```python
## lambda with multiple arguments

area = lambda lenght, width : lenght * width
```

```python
area(5, 3)
```

```python
area(10, 4)
```

```python
# lambda with conditional

is_even = lambda sample: "Even" if sample % 2 == 0 else "Odd"
```

```python
def to_upper(s):
    return s.upper()
```

```python
to_upper("aldkfjadljf")
```

```python
# same using lambda to convert uppercase

to_upper = lambda s: s.upper()

to_upper("ksdjfosjidf")
```

```python
to_lower = lambda s : s.lower()

to_lower("UEHRGIUEHRG")
```

```python
is_even(9)
```

```python
is_even(10)
```

```python
def reverse_string(s):
    return s[::-1]
```

```python
reverse_string("reverse this")
```

```python
reversed_string = lambda s : s[::-1]
reversed_string("hello")
```

```python
def greet(name: str, age: int) -> str:
    """paramter
    name : str
    age: int

    return str"""
    return f"Hello, {name}! you are {age} year old"
```

```python
## Basic decorator structure

def decorator_function(orginal_function):

    def wrapper():
        print("Before fucntion runs")
        orginal_function()
        print("After function runs")

    return wrapper
```

```python
decorated = decorator_function(greet)
```

```python
decorated()
```
