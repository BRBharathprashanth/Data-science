```python
def complex_function(a, b, *args, c = 10, **kwargs):
    print(f"a (position): {a}")
    print(f"b (position): {b}")
    print(f"args (variable position): {args}")
    print(f"c (default): {c}")
    print(f"*kwargs (vairblae keyword): {kwargs}")
```

```python
complex_function(1,2,3,4,5, c=20, x=100, y = 200)
```

```python
####multiple return values
def calculate_stats(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    average = sum(numbers)/ len(numbers)

    return minimum, maximum, average
```

```python
min_val, max_val, avg_val = calculate_stats([10,20,30,40,50,60])
```

```python
min_val, max_val, avg_val
```

```python
###return without value
def print_report(data):
    if not data:
        print("No data to display")
        return # Exit the fucntion early

    print("Data Report: ")
    for item in data:
        print(f" . {item}")
```

```python
print_report([])
```

```python
####cal mean, mode, median
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)

    if n % 2 == 0:
        mid1 = sorted_numbers[n//2 - 1]
        mid2 = sorted_numbers[n//2]
        return (mid1 + mid2)/ 2
    else:
        return sorted_numbers[n//2]
```

```python
##median
data = [23, 23, 23, 23, 34, 43, 44, 45, 45, 56, 67, 67, 67, 89]

median_of_nums = calculate_median(data)
print(median_of_nums)
```

```python
##mean
data = [23, 23, 23, 23, 34, 43, 44, 45, 45, 56, 67, 67, 67, 89]

mean_of_nums = calculate_mean(data)
print(mean_of_nums)
```

```python
###mode
def calculate_mode(numbers):
    counts = Counter(numbers)
    max_count = max(counts.values())
    modes = [num for num, count in counts.items() if count == max_count]
    return modes

mode_result = calculate_mode([23, 23, 23, 23, 34, 43, 44, 45, 56, 67, 67, 67, 89])
print(mode_result)
```
