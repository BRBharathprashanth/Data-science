```python
# List of numbers square if positive , cube if negative
nums = [-3,-2,-1,0,1,2,3]
result = []
for num in nums:
    if num > 0:
        result.append(num ** 2)
    else:
        result.append(num ** 3)
print(result)
```

```python
result = [num**2 if num>0 else num**3 for num in nums]
result
```

```python
##Write a program to find all even numbers greater than 4.

numbers = [1, 2, 3, 4, 5, 6, 7, 8]
result = [x for x in numbers if x % 2 == 0 if x > 4]
result
```

```python
##Write a program to find all odd numbers less than 6.

numbers = [1, 2, 3, 4, 5, 6, 7]
result = [x for x in numbers if x % 2 != 0 if x < 6]
result
```

```python
##Write a program to find numbers divisible by both 2 and 3.

numbers = [1, 2, 3, 4, 5, 6, 12]
result = [x for x in numbers if x % 2 == 0 if x % 3 == 0]
result
```

```python
##Write a program to label numbers as "even" or "odd".

numbers = [1, 2, 3, 4, 5]
result = ["even" if x % 2 == 0 else "odd" for x in numbers]
result
```

```python
##Write a program to square even numbers and keep odd numbers unchanged.

numbers = [1, 2, 3, 4]
result = [x**2 if x % 2 == 0 else x for x in numbers]
result
```

```python
##Write a program to replace negative numbers with zero.

values = [-3, -1, 0, 2, 4]
result = [x if x >= 0 else 0 for x in values]
result
```
