Question 15: Advanced Array Operations

```python
##Q15.1: Apply a function to calculate 5% interest on all account balances

interest_rate = 0.05
account_balances_with_interest = banking_data[:, 1] * (1 + interest_rate)
print(account_balances_with_interest)
```

```python
##Q15.2: Use np.where to categorize customers by balance ranges

balance_ranges = np.where(banking_data[:, 1] < 210, 'Low', np.where(banking_data[:, 1] < 240, 'Medium', 'High'))
print(balance_ranges)
```

```python
##Q15.3: Use np.select to categorize customers by credit score ranges
import numpy as np
credit_scores = np.array([520, 610, 690, 720, 810])

conditions = [
    credit_scores < 580,
    (credit_scores >= 580) & (credit_scores < 670),
    (credit_scores >= 670) & (credit_scores < 740),
    credit_scores >= 740
]

# Categories
choices = ['Poor', 'Fair', 'Good', 'Excellent']

# Apply np.select with an explicit string default value
credit_category = np.select(conditions, choices, default='Unknown')

print(credit_category)
```

```python
##Q15.4: Use np.piecewise to apply different interest rates based on balance

banking_data = np.array([[100, 200, 300], [150, 250, 350], [120, 220, 320]])

conditions = [
    banking_data[:, 1] < 210,
    (banking_data[:, 1] >= 210) & (banking_data[:, 1] < 240),
    banking_data[:, 1] >= 240
]

functions = [
    lambda x: x * (1 + 0.01),
    lambda x: x * (1 + 0.02),
    lambda x: x * (1 + 0.03)
]

account_balances_with_piecewise_interest = np.piecewise(banking_data[:, 1], conditions, functions)
print(account_balances_with_piecewise_interest)
```

Question 16: Array Concatenation and Stacking

```python
##Q16.1: Concatenate two arrays horizontally using np.hstack()


array1_to_stack = np.array([[1, 2], [3, 4], [5, 6]])
array2_to_stack = np.array([[7], [8], [9]])

stacked_array = np.hstack((array1_to_stack, array2_to_stack))
print(stacked_array)
```

```python
##Q16.2: Concatenate two arrays vertically using np.vstack()

array1_to_stack = np.array([[1, 2], [3, 4], [5, 6]])
array2_to_stack = np.array([[7, 8], [9, 10], [11, 12]])

stacked_array = np.vstack((array1_to_stack, array2_to_stack))
print(stacked_array)
```

```python
##Q16.3: Concatenate arrays along a specific axis using np.concatenate()

array1_to_stack = np.array([[1, 2], [3, 4], [5, 6]])
array2_to_stack = np.array([[7,8], [9, 10], [11, 12]])

stacked_array = np.concatenate((array1_to_stack, array2_to_stack), axis=1)
print(stacked_array)
```

```python
##Q16.4: Stack arrays depth-wise using np.dstack()

array1_dstack = np.array([[1, 2], [3, 4]])
array2_dstack = np.array([[5, 6], [7, 8]])

stacked_dstack = np.dstack((array1_dstack, array2_dstack))
print(stacked_dstack)
```

```python
##Q16.5: Create a column stack using np.column_stack()

array1_column = np.array([1, 2, 3])
array2_column = np.array([4, 5, 6])

column_stacked_array = np.column_stack((array1_column, array2_column))
print(column_stacked_array)
```
