```python
##Q: Write a Python program to count the frequency of each character in a given string.
s="interview"
freq={}
for i in s:
    freq[i]=freq.get(i,0)+1
print(freq)
```

```python
##Q: Write a Python program to count the frequency of each word in a given sentence
s = "data science is data driven and data science is fun"
freq={}
words = s.split()
freq={}
print(words)
for word in words:
    freq[word] = freq.get(word,0)+1
print(freq)
```
