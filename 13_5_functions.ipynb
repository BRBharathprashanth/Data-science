```python
##accuracy
models = [
    {"name": "RandomForest", "accuracy": 0.89},
    {"name": "XGBoost", "accuracy": 0.92},
    {"name": "SVM", "accuracy": 0.85}

]

sorted(models,key=lambda x : x['accuracy'])
```

```python
##accuracy decrease function

models = [
    {"name": "RandomForest", "accuracy": 0.89},
    {"name": "XGBoost", "accuracy": 0.92},
    {"name": "SVM", "accuracy": 0.85}

]

sorted(models,key=lambda x : x['accuracy'],reverse= True)
```

```python
## Basic decorator structure

def decorator_function(orginal_function):

    def wrapper():
        print("Before fucntion runs")
        orginal_function()
        print("After function runs")
    return wrapper

def greet():
    print("Hello!")

result  = decorator_function(greet)
result()
```
