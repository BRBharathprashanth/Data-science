```python
##Calculate the sum of digits in number 12345
num = 12345
sum = 0

while num > 0:
    digit = num % 10
    sum += digit
    num = num // 10

print(f"The sum of the digits is: {sum}")
```

```python
##check if 97 is a prime number
num = 97
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```
